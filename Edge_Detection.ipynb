# 1. Import Dependencies

```python
import cv2
import os
```

# 2. Load Image

```python
IMG_PATH = os.path.join('/content/540654.jpg')
```

```python
img = cv2.imread(IMG_PATH)
```

# 3. Edge Detection

```python
#https://learnopencv.com/edge-detection-using-opencv/
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(img, (5,5),0)
```

```python
canny = cv2.Canny(blurred, threshold1=180, threshold2=200)
```

```python
from google.colab.patches import cv2_imshow
resized = cv2.resize(canny, (int(img.shape[1]/2), int(img.shape[0]/2)))
cv2_imshow(resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 4. Export Edges

```python
import numpy as np
```

```python
np.savetxt('edges.csv',canny, delimiter=",")
```

```python
cv2.imwrite('edges.jpg', resized)
```
